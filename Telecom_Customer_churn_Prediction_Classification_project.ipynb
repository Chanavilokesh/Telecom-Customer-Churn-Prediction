banking, and SaaS. Losing customers leads to revenue loss and increased acquisition costs.

● Customer Age

● Gender

● Tenure

● Total Charges

● Payment Method

● Internet Service

● Tech Support

● Online Security

● Customer Support Calls

Students are required to analyze the dataset, preprocess the data, perform feature engineering,

## **Data Description**
The dataset contains **customer information** from a telecom company, including demographics, service subscriptions, account details, billing information, and a **churn flag** indicating whether a customer has left the service. It helps analyze **customer behavior** and identify key factors contributing to **customer churn**.

**customerID** – Unique identifier assigned to each customer.

**gender** – Customer’s gender (Male/Female).

**SeniorCitizen** – Indicates if the customer is a senior citizen (0 = No, 1 = Yes).

**Partner** – Whether the customer has a partner (Yes/No).

**Dependents** – Whether the customer has dependents (Yes/No).

**tenure** – Number of months the customer has stayed with the company.

**PhoneService** – Indicates if the customer has phone service (Yes/No).

**MultipleLines** – Availability of multiple phone lines.

**InternetService** – Type of internet service (DSL/Fiber optic/No).

**OnlineSecurity** – Online security service status.

**OnlineBackup** – Online backup service status.

**DeviceProtection** – Device protection plan status.

**TechSupport** – Technical support service status.

**StreamingTV** – Access to streaming TV service.

**StreamingMovies** – Access to streaming movie service.

**Contract** – Customer’s contract type (Month-to-month/One year/Two year).

**PaperlessBilling** – Indicates if customer uses paperless billing (Yes/No).

**PaymentMethod** – Customer’s mode of payment.

**MonthlyCharges** – Monthly charges billed to the customer.

**TotalCharges** – Total charges incurred by the customer.

**Churn** – Indicates whether the customer has discontinued the service (Yes/No).

# **Key Drivers of Customer Churn**

Customers who are charged more every month tend to feel the service is expensive.
This increases dissatisfaction and leads to a higher chance of leaving the service.


Customers with short-term contracts have more flexibility to switch providers anytime.
This makes them more likely to churn compared to long-term contract customers.


Customers without proper technical support may face unresolved service issues.
This reduces customer satisfaction and increases the probability of churn.


New customers who have recently joined are less attached to the company.
They are more likely to leave early if their expectations are not met.

# **Strategic Recommendations**


● **Offer long-term contract discounts**
Providing discounts and special benefits for customers who choose long-term contracts can encourage them to stay with the company for a longer period and reduce churn rates.



● **Improve customer support services**
Enhancing the quality and responsiveness of customer support helps in resolving issues quickly, leading to higher customer satisfaction and lower chances of customers leaving the service.



● **Provide personalized retention offers**
Creating customized offers based on customer usage and preferences makes customers feel valued, which increases their engagement and reduces the likelihood of churn.



● **Identify high-risk customers early**
Using data analysis to detect customers who are likely to churn allows the company to take proactive measures such as special offers or support to retain them.



● **Reduce churn through loyalty programs**
Implementing loyalty programs that reward long-term customers with discounts or benefits helps in building strong relationships and encourages customers to continue using the service.
